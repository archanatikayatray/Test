##### Employee Salaries

Data obtained for each employee included current salary(Y), a coded variable indicating highest academic degree obtained (1= bachelor's degree, 2= master's degree, 3= doctoral degree), years of experience since last degree ($X_3$), and the number of persons currently supervised ($X_4$). 

(a) Rewrite the data in the form of ($Y_i$, $x_{i1}$, $x_{i2}$,  $x_{i3}$,  $x_{i4}$) for i=1,...,65. 

```r
salary0 <-read.table(file = "http://www.isye.gatech.edu/~ymei/6414/Handouts/salarydata.csv", sep=",");
salary <- salary0; 
salary$V2 <- I(salary0$V2 ==2);
salary$V3 <- I(salary0$V2 ==3);
salary[,4:5] <- salary0[,3:4];
colnames(salary) <- c("Y", "X1", "X2", "X3", "X4"); 
Y_hat <- salary[,1]; 
salary
```

(b) Regress Y on $X_1$, $X_2$, $X_3$, $X_4$ (all at once), using first-order model and ordinary least squares. Obtain the residuals, and plot them against $\hat{Y}$. What does the residual plot mean?

$\textbf{Obtaining Residuals: }$

```r
fm1 <- lm (salary[,1] ~ salary[,2]+salary[,3]+salary[,4]+salary[,5])
fm1
resid(fm1)
```

$\textbf{Plotting residuals:}$

```r
plot(Y_hat, resid(fm1),xlab="Current Salary(Y_hat)", ylab="Residual")
```

What does this residual plot Suggest????? 

(c) Plot the absolute residuals against $X_3$ and against $X_4$. What do these plots suggest about the relation between the standard deviation of the error term and $X_3$ and $X_4$? 

```r
plot (salary[,4], abs(resid(fm1)), xlab="Years of experience since last degree",ylab="Absolute residual" )
plot (salary[,5], abs(resid(fm1)), xlab="Number of persons currently supervised",ylab="Absolute residual" )
```

Write about standard deviation of error term............

(d) Estimate the standard deviation function by regressing the absolute residuals against $X_3$ and $X_4$ (in first-order form), and then calculate the estimated weight for each case using $w_{i} = 1/{\hat{s_i}}^2$, where $\hat{s_i}$ is fitted value for the absolute residuals. Report value of $w_i$ for four i=1,2,64,65. 

$\textbf{Standard deviation function for |residual| against $X_3$}$

```r
reslm <- lm(abs (resid(fm1)) ~ salary[,4]); 
reslm
S1= 1.9603 + 0.5741 * salary[,4]; 
```

The standard deviation function, $S_1$ (absolute residuals, $X_3$)= $\textbf{1.9603 + 0.5741 $X_3$}$


Obtaining the weights: $w_{i} = 1/{\hat{s_i}}^2$

```r
w= 1/ S1^2;
lm2 <- lm(Y_hat ~ salary[,4], weight=w);
summary(lm2)
```

```r
weight1 <- 1.9603 + 0.5741 * salary[1,4]
weight2 <- 1.9603 + 0.5741 * salary[2,4]
weight64 <- 1.9603 + 0.5741 * salary[64,4]
weight65 <- 1.9603 + 0.5741 * salary[65,4]
weight1
weight2
weight64
weight65
```

$\textbf{Standard deviation function for |residual| against $X_4$}$

```r
reslm2 <- lm(abs (resid(fm1)) ~ salary[,5]); 
reslm2
S2= 4.7422 + 0.5187 * salary[,5]; 
```

The standard deviation function, $S_2$ (absolute residuals, $X_3$)= $\textbf{4.7422 + 0.5187 $X_4$}$

```r
w_k= 1/ S2^2;
lm3 <- lm(Y_hat ~ salary[,5], weight=w_k);
summary(lm3)
```
